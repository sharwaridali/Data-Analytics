```python
x = df.iloc[:, [1, 2, 3, 4]].values
print(x)
```

```python
from sklearn.cluster import KMeans
#code to determine the number of clusters. WCSS- Within Cluster Sum of Sqaures
wcss = []   #the number of clusters is decided based on this

for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
    
# Plotting the results onto a line graph, 
plt.plot(range(1, 11), wcss,'go-')
plt.title('The elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS') # Within cluster sum of squares
plt.show()
```

As WCSS doesn't decrease significantly from the number "3" we choose "3" as the umber of clusters

```python
kmeans = KMeans(n_clusters = 3, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
y = kmeans.fit_predict(x)
print(y)
# Visualising the clusters - On the first two columns
plt.scatter(x[y == 0, 0], x[y == 0, 1], s = 50, c = 'blue', label = 'Iris-setosa')
plt.scatter(x[y == 1, 0], x[y == 1, 1], s = 50, c = 'green', label = 'Iris-versicolour')
plt.scatter(x[y == 2, 0], x[y == 2, 1], s = 50, c = 'purple', label = 'Iris-virginica')

# Plotting the centroids of the clusters
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1], s = 100, c = 'yellow', label = 'Centroids')

plt.legend(loc='upper right')
```
