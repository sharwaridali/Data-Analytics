```python
df = pd.DataFrame.drop_duplicates(df)
df.shape
```

```python
df = df.drop(['Country'], axis = 1)
df = df.drop(['Postal Code'], axis = 1)
```

```python
df.info()
```

```python
#to find the total sales
print("Total Sales:" + str(round(sum(df['Sales']), 2)))
#total quantity sold
print("Total Quantity:" + str(sum(df['Quantity'])))
#total profit
print("Total Profit:" + str(round(sum(df['Profit']))))
```

```python
df.describe()
```

```python
corr = df.corr()
print(corr)
sb.heatmap(corr, cmap = 'Purples', annot=True)
```

* Sales and Profit are related upto 48%.
* Quantity and Profit are somewhat related.
* Discount and Profit are negetively related 

```python
sb.pairplot(df, hue='Ship Mode')
```

<h4> Ship Mode <\h4>

```python
print(df['Ship Mode'].value_counts())
sb.countplot(x=df['Ship Mode']) #visualising the count values through graph
```

```python
df.groupby('Ship Mode')[['Profit', 'Sales']].sum().plot.bar(color=['cyan','green'])
```

Conclusion: When the ship mode is Standard Class the Sales and Profit are Maximum,
            When the ship mode is Same day the Sales and Profits are the least.

<h4> Category

```python
print(df['Category'].value_counts())
tp = {'fontsize': 12}
plt.title("Category")
#piechart is in matplotlib
plt.pie(df['Category'].value_counts(), labels = df['Category'].value_counts().index, autopct='%1.2f%%', textprops=tp)
plt.show()
```

```python
df.groupby('Category')[['Profit', 'Sales']].sum().plot.bar(color=['pink', 'purple'])
```

Conclusion: The Sales and Profit for the Technology Category are the Highest

<h4> Sub-Category

```python
print(df['Sub-Category'].value_counts())
df.groupby('Sub-Category')[['Profit', 'Sales']].sum().plot.bar(color=['purple', 'gray'], figsize=(10, 5))
```

```python
statewise = df.groupby(['Sub-Category'])['Profit'].sum().nlargest(50)
statewise.plot.barh(figsize=(10,8))
```

Conclusion: Copiers have the Highest profit

<h4> State

```python
plt.figure(figsize=(18,10))
sb.countplot(x="State", data=df, palette='pastel', order=df['State'].value_counts().index)
plt.xticks(rotation=90)
plt.show()
```

California has the highest number of buyers followed by New York

```python
df.groupby('State')[['Profit', 'Sales']].sum().plot.bar(figsize=(20,10))
```

Conclusion: California and New York have the highest Sales and Profits

<h4> Segment

```python
#segment wise analysis
print(df['Segment'].value_counts())
df.groupby('Segment')[['Profit','Sales']].sum().plot.bar(color=['yellow','black'], figsize=(8,5))
plt.ylabel('Profit/Loss and sales')
plt.show()
```

Conclusion: Sales and Profit are Maximum for Consumer

<h4>Region

```python
print(df['Region'].value_counts())
df.groupby('Region')[['Profit','Sales']].sum().plot.bar(color=['lavender','red'], figsize=(8,5))
plt.show()
```

Conclusion: Profit and Sales are High in the West region

<h4> Region v/s Ship Mode

```python
plt.figure(figsize=(8,5))
sb.countplot(x='Region',hue='Ship Mode',data=df)
plt.ylabel('Count of Ship Mode')
plt.title('Count of ship mode by region and ship mode')
```

Conclusion: Standard Class Ship Mode is preffered by most Customers

<h4> Profit v/s Discount

```python
df.groupby('Discount')['Profit'].sum().plot.barh()
```

Conclusion: There isn't a clear relation in Discount and Profits but when the discounts are low the profits are more
